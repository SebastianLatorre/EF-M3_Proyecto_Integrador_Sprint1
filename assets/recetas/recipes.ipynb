# Generar JSON con recetas
Se implementa notebook colab para generar lista de recetas

```python
import requests
import json
import random
import time

def get_random_meal():
    url = "https://www.themealdb.com/api/json/v1/1/random.php"
    resp = requests.get(url).json()
    return resp.get("meals", [None])[0]

recipes = []
ids_seen = set()

while len(recipes) < 50:
    meal = get_random_meal()
    if not meal: 
        continue
    idm = meal["idMeal"]
    if idm in ids_seen:
        continue
    ids_seen.add(idm)

    # Recoger ingredientes
    ingredients = []
    for i in range(1, 21):
        ing = meal.get(f"strIngredient{i}")
        measure = meal.get(f"strMeasure{i}")
        if ing and ing.strip() != "":
            ingredients.append(f"{measure.strip()} {ing.strip()}".strip())

    recipes.append({
        "id": idm,
        "nombre": meal.get("strMeal"),
        "categoria": meal.get("strCategory"),
        "area": meal.get("strArea"),
        "ingredientes": ingredients,
        "instrucciones": meal.get("strInstructions"),
        "imagen": meal.get("strMealThumb")
    })
    # opcional: evitar saturar la API
    time.sleep(0.2)

with open("recetas_50.json", "w", encoding="utf-8") as f:
    json.dump(recipes, f, ensure_ascii=False, indent=2)

print("Generadas:", len(recipes))
```
